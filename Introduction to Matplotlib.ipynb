## Which on eshould you use pyplot or OO matplotlib

when plotting something quickly it is okay to use pyplot
when plotting something advanced use the OO method

```python
over_50=heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
over_50.plot(kind="scatter",x="age",y="chol",c="target");
```

```python
heart_disease
```

```python
fig,ax=plt.subplots(figsize=(10,5))
over_50.plot(kind="scatter",x="age",y="chol",c="target",ax=ax);
```

```python
#oo method mixed with pyplot
fig,ax=plt.subplots(figsize=(10,5))
over_50.plot(kind="scatter",x="age",y="chol",c="target",ax=ax);
ax.set_xlim([45,100])
```

```python
over_50["target"].values
```

```python
#oo method from scratch

fig,ax=plt.subplots(figsize=(10,6))
#plot the data

scatter=ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])

#customise the plot
ax.set(title="Heart diseases and cholestrol levels across ages",xlabel="Age",ylabel="Cholestrol")

#add a legend
ax.legend(*scatter.legend_elements(),title="target")

#add a horizontal line

ax.axhline(over_50["chol"].mean(),linestyle="--");
```

```python
   #subplot of chol,thalach,age

fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)

#add data to ax0

scatter=ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])
#customise the plot
ax0.set(title="Heart diseases and cholestrol levels across ages ax0",ylabel="Cholestrol")
#add a legend
ax0.legend(*scatter.legend_elements(),title="Target")

#add a mean line

ax0.axhline(over_50["chol"].mean(),linestyle="--");

scatter=ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"])
#customise the plot
ax1.set(title="Heart diseases and cholestrol levels across ages ax1",ylabel="Max heart Rate")
#add a legend
ax1.legend(*scatter.legend_elements(),title="Target")

#add a mean line

ax1.axhline(over_50["thalach"].mean(),linestyle="--");


#add atitle to the figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```

```python
#oo method from scratch

plt.style.use("seaborn-v0_8-whitegrid")
fig,ax=plt.subplots(figsize=(10,6))
#plot the data

scatter=ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"],cmap="winter")#this chnges the color of scattered plot

#customise the plot
ax.set(title="Heart diseases and cholestrol levels across ages",xlabel="Age",ylabel="Cholestrol")

#add a legend
ax.legend(*scatter.legend_elements(),title="target")

#add a horizontal line

ax.axhline(over_50["chol"].mean(),linestyle="--");
```

```python
#customising y and x axis limitaitions

   #subplot of chol,thalach,age

fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)

#add data to ax0

scatter=ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"],cmap="plasma")
#customise the plot
ax0.set(title="Heart diseases and cholestrol levels across ages ax0",ylabel="Cholestrol")

#chnge the axis x limits

ax0.set_xlim([50,80])
ax0.set_ylim([200,500])
#add a legend
ax0.legend(*scatter.legend_elements(),title="Target")

#add a mean line

ax0.axhline(over_50["chol"].mean(),linestyle="--");

scatter=ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"])
#customise the plot
ax1.set(title="Heart diseases and cholestrol levels across ages ax1",ylabel="Max heart Rate")

#customising ax1 x axis limits

ax1.set_xlim([50,80])
ax1.set_ylim([60,200])
#add a legend
ax1.legend(*scatter.legend_elements(),title="Target")

#add a mean line

ax1.axhline(over_50["thalach"].mean(),linestyle="--");


#add atitle to the figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```
